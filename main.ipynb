# สัดส่วนภาษีแต่ละประเภท

```python
tax_map_eng = {
    1: 'PIT',
    2: 'CIT',
    3: 'Petroleum',
    5: 'VAT',
    6: 'SBT',
    7: 'Stamp Duty',
    70: 'Inheritance',
    999: 'Others'
}

df_tax_collect_prep['tax_name'] = df_tax_collect_prep['tax_type'].map(tax_map_eng).fillna('Others')
summary = df_tax_collect_prep.groupby('tax_name')['amt_rcv'].sum()
summary_pct = (summary / summary.sum() * 100).sort_values(ascending=True)

bar_colors = ['#dd8452' if tax == 'PIT' else '#4c72b0' for tax in summary_pct.index]

df_pit = df_tax_collect_prep[df_tax_collect_prep['tax_type'] == 1]
pit_summary = df_pit.groupby('form_code')['amt_rcv'].sum()
pit_summary_pct = (pit_summary / pit_summary.sum() * 100).sort_values(ascending=True)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

summary_pct.plot(kind='barh', color=bar_colors, ax=ax1)
ax1.set_title('Percentage of Tax Collected by Type', fontsize=14)
ax1.set_xlabel('Percentage (%)', fontsize=12)
ax1.set_ylabel('Tax Type', fontsize=12)
ax1.set_xlim(0, summary_pct.max() + 5)

for i, v in enumerate(summary_pct):
    weight = 'bold' if summary_pct.index[i] == 'PIT' else 'normal'
    ax1.text(v + 0.5, i, f"{v:.2f}%", va='center', ha='left', fontsize=10, fontweight=weight)

pit_summary_pct.plot(kind='barh', color='mediumseagreen', ax=ax2)
ax2.set_title('Percentage of PIT Collected by Form Code', fontsize=14)
ax2.set_xlabel('Percentage (%)', fontsize=12)
ax2.set_ylabel('Form Code', fontsize=12)
ax2.set_xlim(0, pit_summary_pct.max() + 5)

for i, v in enumerate(pit_summary_pct):
    ax2.text(v + 0.5, i, f"{v:.2f}%", va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.show()
```

# โครงสร้างประชากรไทยเปลี่ยนแปลงอย่างไรในช่วงปี 2020–2025 และสัดส่วนวัยทำงานลดลงหรือไม่?

```python
pop_df = pd.read_excel(file_path, sheet_name="ประชากร")
pop_df.head()
```

```python
plt.rcParams['font.size'] = 12
pop_df['Value'] = pd.to_numeric(pop_df['Value'], errors='coerce').fillna(0)

# Group Age Categories
def map_age_group(gen):
    if gen == '0_0-15':
        return 'Children (0-15)'
    elif gen == '11_morethanequal60':
        return 'Elderly (60+)'
    else:
        return 'Working Age (15-59)'

pop_df['Age_Group'] = pop_df['Generation'].apply(map_age_group)

pop_yearly = pop_df.groupby(['year', 'Age_Group'])['Value'].sum().unstack()
cols = ['Children (0-15)', 'Working Age (15-59)', 'Elderly (60+)']
pop_yearly = pop_yearly[cols]
pop_yearly.loc[2025] = (pop_yearly.loc[2025] / 2).astype(int)
pop_yearly_pct = pop_yearly.div(pop_yearly.sum(axis=1), axis=0) * 100

tbl_pop_yearly = pop_yearly
tbl_pop_yearly['Total'] = pop_yearly['Children (0-15)'] + pop_yearly['Working Age (15-59)'] + pop_yearly['Elderly (60+)']

tbl_pop_yearly
```

```python
def millions_formatter(x, pos):
    if x == 0:
        return '0'
    return f'{int(x / 1e6)}m'

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

pop_yearly_pct.plot(kind='bar', stacked=True, color=['#FF9999', '#66B2FF', '#99FF99'], alpha=0.9, ax=ax1, edgecolor='white')

for idx, year in enumerate(pop_yearly_pct.index):
    y_vals = pop_yearly_pct.loc[year]
    y1 = y_vals['Children (0-15)']
    y2 = y_vals['Working Age (15-59)']
    y3 = y_vals['Elderly (60+)']

    mid1 = y1 / 2
    mid2 = y1 + (y2 / 2)
    mid3 = y1 + y2 + (y3 / 2)

    ax1.text(idx, mid1, f"{y1:.1f}%", ha='center', va='center', fontsize=10, color='black')
    ax1.text(idx, mid2, f"{y2:.1f}%", ha='center', va='center', fontsize=10, color='black')
    ax1.text(idx, mid3, f"{y3:.1f}%", ha='center', va='center', fontsize=10, color='black')

ax1.set_title('Population Structure (%)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Percentage (%)')
ax1.set_ylim(0, 105)
ax1.set_xticklabels(pop_yearly_pct.index, rotation=0)
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)

years_str = pop_yearly_pct.index.astype(str)

ax2.plot(years_str, pop_yearly['Children (0-15)'], color='#FF9999', marker='o', linewidth=2, label='Children (0-15)')
ax2.plot(years_str, pop_yearly['Working Age (15-59)'], color='#66B2FF', marker='o', linewidth=2, label='Working Age (15-59)')
ax2.plot(years_str, pop_yearly['Elderly (60+)'], color='#99FF99', marker='o', linewidth=2, label='Elderly (60+)')

ax2.set_title('Population Trend (Count)')
ax2.set_xlabel('Year')
ax2.set_ylabel('Population Count')
ax2.set_ylim(bottom=0)
ax2.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)
ax2.grid(axis='y', linestyle='--', alpha=0.7)
ax2.yaxis.set_major_formatter(ticker.FuncFormatter(millions_formatter))

plt.tight_layout()
plt.show()
```

# ภาษีเงินได้บุคคลธรรมดาต่อแรงงาน 1 คน (Tax per Worker) มีแนวโน้มเพิ่มขึ้นหรือลดลง?

```python
labor_df = pd.read_excel(file_path, sheet_name="แรงงาน")
labor_df.head()
```

```python
ss_df = pd.read_excel(file_path, sheet_name="social_security_prep")
ss_df.head()
```

```python
tax_df = pd.read_excel(file_path, sheet_name="tax_collect_prep")
tax_df.head()
```

```python
workers_df = labor_df[labor_df['labor_force_status'] == 'กำลังแรงงานปัจจุบัน_ผู้มีงานทำ']
total_workers = workers_df.groupby(['year', 'province'])['value'].mean().groupby('year').sum().reset_index(name='Total_Employed')

ss_df = ss_df[['year', 'article_33']]

headcount_df = pd.merge(total_workers, ss_df, on='year', how='inner')
headcount_df['Salary_Headcount'] = headcount_df['article_33']
headcount_df['Non_Salary_Headcount'] = headcount_df['Total_Employed'] - headcount_df['article_33']

headcount_df['Salary_Pct'] = (headcount_df['Salary_Headcount'] / headcount_df['Total_Employed']) * 100
headcount_df['Non_Salary_Pct'] = (headcount_df['Non_Salary_Headcount'] / headcount_df['Total_Employed']) * 100

headcount_df = headcount_df[['year', 'Salary_Headcount', 'Non_Salary_Headcount', 'Total_Employed', 'Salary_Pct', 'Non_Salary_Pct']]

tax_df['amt_rcv'] = pd.to_numeric(tax_df['amt_rcv'], errors='coerce').fillna(0)
pit_df = tax_df[tax_df['tax_type'] == 1].copy()

def classify_tax_payer(form):
    form = str(form).strip()
    if form in ['ภงด1', 'ภงด91']:
        return 'Salary_Tax'
    elif form in ['ภงด2', 'ภงด3', 'ภงด90', 'ป2', 'ทป3']:
        return 'Non_Salary_Tax'
    else:
        return 'Other'

pit_df['Payer_Type'] = pit_df['form_code'].apply(classify_tax_payer)
tax_grouped = pit_df.groupby(['year', 'Payer_Type'])['amt_rcv'].sum().unstack().reset_index()

total_tax_df = pit_df.groupby('year')['amt_rcv'].sum().reset_index(name='Total_Tax')
tax_grouped = pd.merge(tax_grouped, total_tax_df, on='year', how='left')

final_df = pd.merge(headcount_df, tax_grouped, on='year', how='inner')
final_df['Tax_Per_Salary_Earner'] = final_df['Salary_Tax'] / final_df['Salary_Headcount']
final_df['Tax_Per_Non_Salary_Earner'] = final_df['Non_Salary_Tax'] / final_df['Non_Salary_Headcount']
final_df['Overall_Average_Tax'] = final_df['Total_Tax'] / final_df['Total_Employed']
final_df['year'] = final_df['year'].astype(int)
```

```python
plt.rcParams['font.size'] = 14
fig, ax = plt.subplots(figsize=(10, 6))

years = ['2021', '2022', '2023', '2024']
non_sal_pct = headcount_df['Non_Salary_Pct']
sal_pct = headcount_df['Salary_Pct']

ax.bar(years, non_sal_pct, label='Non-Salary / Others', color='#dd8452', edgecolor='white', width=0.6)
ax.bar(years, sal_pct, bottom=non_sal_pct, label='Salary Earners', color='#4c72b0', edgecolor='white', width=0.6)
ax.set_title('Proportion of Workforce: Salary vs Non-Salary Earners (2021-2024)', pad=15)
ax.set_ylabel('Percentage of Workforce (%)')
ax.set_xlabel('Year')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

for idx, year in enumerate(headcount_df.index):
    n_pct = headcount_df.loc[year, 'Non_Salary_Pct']
    s_pct = headcount_df.loc[year, 'Salary_Pct']

    n_val = headcount_df.loc[year, 'Non_Salary_Headcount'] / 1e6
    s_val = headcount_df.loc[year, 'Salary_Headcount'] / 1e6

    mid_n = n_pct / 2
    mid_s = n_pct + (s_pct / 2)

    ax.text(idx, mid_n, f'{n_pct:.1f}%\n({n_val:.1f}M)', ha='center', va='center', color='black')
    ax.text(idx, mid_s, f'{s_pct:.1f}%\n({s_val:.1f}M)', ha='center', va='center', color='white')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(final_df['year'], final_df['Tax_Per_Salary_Earner'], marker='o', markersize=8, color='#4c72b0', linewidth=3, label='Salary Earners (PND 1, 91)')
ax.plot(final_df['year'], final_df['Tax_Per_Non_Salary_Earner'], marker='s', markersize=8, color='#dd8452', linewidth=3, label='Non-Salary / Freelance')
ax.plot(final_df['year'], final_df['Overall_Average_Tax'], marker='^', markersize=8, color='#555555', linewidth=3, linestyle='--', label='Overall Average (All Workers)')

for i, row in final_df.iterrows():
    ax.annotate(f"{row['Tax_Per_Salary_Earner']:,.0f}",
                xy=(row['year'], row['Tax_Per_Salary_Earner']),
                xytext=(0, 10), textcoords="offset points",
                ha='center', va='bottom', color='#4c72b0', fontweight='bold')

    ax.annotate(f"{row['Tax_Per_Non_Salary_Earner']:,.0f}",
                xy=(row['year'], row['Tax_Per_Non_Salary_Earner']),
                xytext=(0, -15), textcoords="offset points",
                ha='center', va='top', color='#dd8452', fontweight='bold')

    ax.annotate(f"{row['Overall_Average_Tax']:,.0f}",
                xy=(row['year'], row['Overall_Average_Tax']),
                xytext=(0, 10), textcoords="offset points",
                ha='center', va='bottom', color='#555555', fontweight='bold')

ax.set_title('Average Personal Income Tax per Person (2022-2024)')
ax.set_xlabel('Year')
ax.set_ylabel('Tax per Person (THB / Year)')
ax.set_xticks(final_df['year'])
ax.set_ylim(0, final_df['Tax_Per_Salary_Earner'].max() * 1.2)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)
ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
