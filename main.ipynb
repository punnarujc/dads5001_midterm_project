```python
plt.rcParams['font.size'] = 12
pop_df['Value'] = pd.to_numeric(pop_df['Value'], errors='coerce').fillna(0)

# Group Age Categories
def map_age_group(gen):
    if gen == '0_0-15':
        return 'Children (0-15)'
    elif gen == '11_morethanequal60':
        return 'Elderly (60+)'
    else:
        return 'Working Age (15-59)'

pop_df['Age_Group'] = pop_df['Generation'].apply(map_age_group)

pop_yearly = pop_df.groupby(['year', 'Age_Group'])['Value'].sum().unstack()
cols = ['Children (0-15)', 'Working Age (15-59)', 'Elderly (60+)']
pop_yearly = pop_yearly[cols]
pop_yearly.loc[2025] = (pop_yearly.loc[2025] / 2).astype(int)
pop_yearly_pct = pop_yearly.div(pop_yearly.sum(axis=1), axis=0) * 100

tbl_pop_yearly = pop_yearly
tbl_pop_yearly['Total'] = pop_yearly['Children (0-15)'] + pop_yearly['Working Age (15-59)'] + pop_yearly['Elderly (60+)']

tbl_pop_yearly
```

# การเปลี่ยนแปลงโครงสร้างประชากรส่งผลต่อภาระและความสามารถในการรองรับของระบบประกันสังคมไทยอย่างไร?

```python
plt.rcParams['font.family'] = 'TH Sarabun New'

# =========================
# (1) GROUP RAW DATA
# =========================
df_gen = (
    df_ประชากร
    .groupby(['year', 'Generation'], as_index=False)['Value']
    .sum()
)

# =========================
# (2) TOTAL (ตัวจริง)
# =========================
df_total_fix = (
    df_ประชากร
    .drop_duplicates(subset=['year','province'])
    .groupby('year')['ผลรวมประชากรทั้งหมด']
    .sum()
)

# =========================
# (3) MAP AGE GROUP
# =========================
def map_age_group(gen):
    if gen == '0_0-15':
        return '0-14'
    elif gen == '11_morethanequal60':
        return '60+'
    else:
        return '15-59'

df_gen['age_group'] = df_gen['Generation'].apply(map_age_group)

# =========================
# (4) GROUP 3 AGE GROUP
# =========================
df_grouped = (
    df_gen
    .groupby(['year', 'age_group'], as_index=False)['Value']
    .sum()
)

# =========================
# (5) CALCULATE SHARE (ใช้ total ที่ fix)
# =========================
df_grouped['total'] = df_grouped['year'].map(df_total_fix)

df_grouped['share'] = df_grouped['Value'] / df_grouped['total']

# =========================
# (6) PIVOT + FIX ORDER
# =========================
pivot_group = (
    df_grouped
    .pivot(index='year', columns='age_group', values='share')
    .reindex(columns=['0-14','15-59','60+'])
    .sort_index()
)


pivot_group = pivot_group.div(pivot_group.sum(axis=1), axis=0)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# =========================
# (1) POPULATION DATA
# =========================
pivot_share = pivot_group.rename(columns={
    '0-14': 'Children',
    '15-59': 'Working',
    '60+': 'Elderly'
})

df_total = (
    df_ประชากร
    .drop_duplicates(subset=['year', 'province'])
    .groupby('year')['ผลรวมประชากรทั้งหมด']
    .sum()
)

pivot_count = pivot_share.mul(df_total, axis=0)

# =========================
# (2) M33 DATA
# =========================
df_m33 = df_age_social_security_prep_article33.copy()

df_m33['group'] = df_m33['Age_Years'].apply(
    lambda x: 'Elderly' if '60' in str(x) else 'Working'
)

df_m33_group = (
    df_m33.groupby(['year', 'group'])['Total']
    .sum()
    .reset_index()
)

pivot_m33 = df_m33_group.pivot(index='year', columns='group', values='Total')

# =========================
# (3) MERGE TABLE
# =========================
df_plot = pd.DataFrame({
    'Year': pivot_count.index.astype(int),
    'Working-age Population': pivot_count['Working'],
    'Elderly Population': pivot_count['Elderly'],
    'M33 Contributors (Working)': pivot_m33['Working'],
    'M33 Elderly Beneficiaries': pivot_m33['Elderly']
}).reset_index(drop=True)

# =========================
# (4) FORMAT TABLE (สำหรับแสดงผล)
# =========================
df_display = df_plot.copy()
for col in df_display.columns[1:]:
    df_display[col] = df_display[col].apply(
        lambda x: f"{int(x):,}" if pd.notnull(x) else "-"
    )

# =========================
# (5) METRICS
# =========================
coverage_working = df_plot['M33 Contributors (Working)'] / df_plot['Working-age Population']
coverage_elderly = df_plot['M33 Elderly Beneficiaries'] / df_plot['Elderly Population']
burden = df_plot['Elderly Population'] / df_plot['Working-age Population']
support_ratio = df_plot['Working-age Population'] / df_plot['Elderly Population']
fiscal_pressure = df_plot['Elderly Population'] / df_plot['M33 Contributors (Working)']
support_capacity = df_plot['M33 Contributors (Working)'] / df_plot['Elderly Population']

years = df_plot['Year']

# =========================
# 🔑 HELPER FUNCTION (สำคัญมาก)
# =========================
def get_last_valid(series, years_series):
    valid = series.dropna()
    last_idx = valid.index[-1]
    return years_series.iloc[last_idx], series.iloc[last_idx], last_idx

# =========================
# (A) FIG 1: TABLE แยกต่างหาก
# =========================
fig_table = plt.figure(figsize=(18, 6))
ax_table = fig_table.add_subplot(111)
ax_table.axis('off')

table = ax_table.table(
    cellText=df_display.values,
    colLabels=df_display.columns,
    loc='center',
    cellLoc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)

# header style
for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_text_props(weight='bold')
        cell.set_facecolor('#eeeeee')

ax_table.set_title(
    "Population vs Social Security (M33) by Age Group",
    fontsize=15,
    weight='bold',
    pad=18
)

plt.tight_layout()
plt.show()

# =========================
# (B) FIG 2: GRAPHS แยกต่างหาก (2x3)
# =========================
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

# ---------- (1) Working Coverage ----------
ax1 = axes[0]
ax1.plot(years, coverage_working, marker='o', color='blue')
ax1.set_title("Working Coverage\n(M33 / Population)")
ax1.set_xticks(years)
ax1.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.0%}"))
ax1.grid(alpha=0.3)

# ---------- (2) Burden ----------
ax2 = axes[1]
ax2.plot(years, burden, marker='o', color='orange')
ax2.set_title("Burden on Working\n(Elderly / Working)")
ax2.set_xticks(years)
ax2.grid(alpha=0.3)

x_last, y_last, _ = get_last_valid(burden, years)
ax2.annotate(
    "More elderly per worker ↑",
    xy=(x_last, y_last),
    xytext=(x_last - 1, y_last * 1.03),
    fontsize=12,
    fontweight='bold',
    arrowprops=dict(arrowstyle="->", lw=2)
)

# ---------- (3) Fiscal Pressure ----------
ax3 = axes[2]
ax3.plot(years, fiscal_pressure, marker='o', color='red')
ax3.set_title("Fiscal Pressure\n(Elderly / M33 Working)")
ax3.set_xticks(years)
ax3.grid(alpha=0.3)

x_last, y_last, _ = get_last_valid(fiscal_pressure, years)
ax3.annotate(
    "Fiscal pressure rising ↑",
    xy=(x_last, y_last),
    xytext=(x_last - 1, y_last * 0.98),
    fontsize=12,
    fontweight='bold',
    color='black',
    arrowprops=dict(arrowstyle="->", color='black', lw=2)
)

# ---------- (4) Elderly Coverage ----------
ax4 = axes[3]
ax4.plot(years, coverage_elderly, marker='o', color='green')
ax4.set_title("Elderly Coverage\n(M33 / Population)")
ax4.set_xticks(years)
ax4.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.1%}"))
ax4.grid(alpha=0.3)

# ---------- (5) Support Ratio ----------
ax5 = axes[4]
ax5.plot(years, support_ratio, marker='o', color='purple')
ax5.set_title("Support Ratio\n(Working / Elderly)")
ax5.set_xticks(years)
ax5.grid(alpha=0.3)

x_last, y_last, _ = get_last_valid(support_ratio, years)
ax5.text(
    x_last,
    y_last,
    "Fewer workers ↓ ",
    fontsize=12,
    weight='bold'
)

# ---------- (6) Support Capacity ----------
ax6 = axes[5]
ax6.plot(years, support_capacity, marker='o', color='brown')
ax6.set_title("Support Capacity\n(M33 Working / Elderly)")
ax6.set_xticks(years)
ax6.grid(alpha=0.3)

x_last, y_last, _ = get_last_valid(support_capacity, years)
ax6.text(
    x_last,
    y_last,
    "Lower capacity ↓ ",
    fontsize=12,
    weight='bold'
)

plt.tight_layout()
plt.savefig('08_Working_Ratio.png')
plt.show()
```

# ภาษีเงินได้บุคคลธรรมดาต่อแรงงาน 1 คน (Tax per Worker) มีแนวโน้มเพิ่มขึ้นหรือลดลง?

```python
labor_df = pd.read_excel(file_path, sheet_name="แรงงาน")
labor_df.head()
```

```python
workers_df = labor_df[labor_df['labor_force_status'] == 'กำลังแรงงานปัจจุบัน_ผู้มีงานทำ']
total_workers = workers_df.groupby(['year', 'province'])['value'].mean().groupby('year').sum().reset_index(name='Total_Employed')

ss_df = ss_df[['year', 'article_33']]

headcount_df = pd.merge(total_workers, ss_df, on='year', how='inner')
headcount_df['Salary_Headcount'] = headcount_df['article_33']
headcount_df['Non_Salary_Headcount'] = headcount_df['Total_Employed'] - headcount_df['article_33']

headcount_df['Salary_Pct'] = (headcount_df['Salary_Headcount'] / headcount_df['Total_Employed']) * 100
headcount_df['Non_Salary_Pct'] = (headcount_df['Non_Salary_Headcount'] / headcount_df['Total_Employed']) * 100

headcount_df = headcount_df[['year', 'Salary_Headcount', 'Non_Salary_Headcount', 'Total_Employed', 'Salary_Pct', 'Non_Salary_Pct']]

tax_df['amt_rcv'] = pd.to_numeric(tax_df['amt_rcv'], errors='coerce').fillna(0)
pit_df = tax_df[tax_df['tax_type'] == 1].copy()

def classify_tax_payer(form):
    form = str(form).strip()
    if form in ['ภงด1', 'ภงด91']:
        return 'Salary_Tax'
    elif form in ['ภงด2', 'ภงด3', 'ภงด90', 'ป2', 'ทป3']:
        return 'Non_Salary_Tax'
    else:
        return 'Other'

pit_df['Payer_Type'] = pit_df['form_code'].apply(classify_tax_payer)
tax_grouped = pit_df.groupby(['year', 'Payer_Type'])['amt_rcv'].sum().unstack().reset_index()

total_tax_df = pit_df.groupby('year')['amt_rcv'].sum().reset_index(name='Total_Tax')
tax_grouped = pd.merge(tax_grouped, total_tax_df, on='year', how='left')

final_df = pd.merge(headcount_df, tax_grouped, on='year', how='inner')
final_df['Tax_Per_Salary_Earner'] = final_df['Salary_Tax'] / final_df['Salary_Headcount']
final_df['Tax_Per_Non_Salary_Earner'] = final_df['Non_Salary_Tax'] / final_df['Non_Salary_Headcount']
final_df['Overall_Average_Tax'] = final_df['Total_Tax'] / final_df['Total_Employed']
final_df['year'] = final_df['year'].astype(int)
```

# คนเสียภาษี ม.33 กำลังแบกคนในประเทศอยู่หรือเปล่า?

**คำถามหลัก:** กลุ่มผู้ประกันตนตามมาตรา 33 (ม.33) ที่จ่ายภาษีและเป็นฐานรายได้ของประเทศ กำลังแบกรับ “ผู้พึ่งพิง” (คนที่ไม่ได้อยู่ในระบบ ม.33) มากแค่ไหน และภาคใดมีแรงกดดันสูงสุด?

**นิยามสั้น:** **ผู้แบก** = ผู้ประกันตน ม.33 | **ผู้พึ่งพิง** = ประชากรที่ไม่ได้อยู่ในระบบ ม.33 (Non_M33)  
**ข้อมูล:** summary_raw_data.xlsx (ประชากร, ม.33 รายภาค/อายุ, ภาษี PIT, กำลังแรงงาน)


```python
# เตรียมข้อมูลสำหรับ 3 ข้อ (ใช้เฉพาะ library จาก cell 1)
# ข้อมูลจาก summary_raw_data: age_social_security_prep, region_social_security_prep, ประชากร, tax_collect_prep, แรงงาน

PALETTE = {'blue': '#4c72b0', 'orange': '#dd8452', 'green': '#55a868', 'gray': '#555555', 'red': '#d62728'}
N_REGIONS = 7  # จำนวนภาคที่แสดงในตาราง/กราฟ
# สไตล์กราฟให้เหมือน reference (bar สีอ่อน, เส้นน้ำเงิน, grid แบบ dashed)
CHART = {'bar_fill': '#ffd1dc', 'bar_edge': '#f8b4c4', 'line_blue': '#0000FF', 'line_dark': '#2c3e50',
         'title_fontsize': 16, 'title_pad': 20, 'label_fontsize': 13, 'grid_alpha': 0.3, 'grid_ls': '--'}

def style_pastel_table(df, highlight_first=True):
    df_disp = df.copy()
    numeric_cols = df_disp.select_dtypes(include=[np.number]).columns.tolist()
    count_keys = ['M33', 'Pop', 'Non_M33', 'Employed', 'PIT']
    ratio_keys = ['Ratio', 'Index']
    fmt = {}
    for c in df_disp.columns:
        s = str(c)
        if any(k in s for k in count_keys):
            fmt[c] = '{:,.0f}'
        elif any(k in s for k in ratio_keys):
            fmt[c] = '{:.4f}'
    sty = df_disp.style.format({k: v for k, v in fmt.items() if k in df_disp.columns})
    sty = sty.set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#f3f0ff'), ('font-weight', '600'), ('border', '1px solid #e6e3ff')]},
        {'selector': 'td', 'props': [('border', '1px solid #eeeeee')]},
    ])
    if numeric_cols:
        sty = sty.background_gradient(subset=[c for c in numeric_cols if c in df_disp.columns], cmap='Pastel1', axis=0)
    return sty

# --- โหลดชีท ---
all_sheets = pd.read_excel(file_path, sheet_name=None)
age_df = all_sheets['age_social_security_prep'].copy()
region_m33 = all_sheets['region_social_security_prep'].copy()
pop_df = all_sheets['ประชากร'].copy()
tax_df = all_sheets['tax_collect_prep'].copy()
labor_df = all_sheets['แรงงาน'].copy()

# --- M33 (ผู้ประกันตน) ---
if 'Age (years)' in age_df.columns:
    age_df = age_df.rename(columns={'Age (years)': 'Age_Years'})
if 'No_group_Age' in age_df.columns:
    age_df = age_df.rename(columns={'No_group_Age': 'Generation'})
age_df['year'] = pd.to_numeric(age_df['year'], errors='coerce')
age_df['Total'] = pd.to_numeric(age_df['Total'], errors='coerce')
age_df = age_df.dropna(subset=['year', 'Total'])

m33_nat = age_df.groupby('year', as_index=False)['Total'].sum().rename(columns={'Total': 'M33_nat'})
m33_region = region_m33.groupby(['year', 'region'], as_index=False)['article33'].sum().rename(columns={'article33': 'M33_region'})

# --- ประชากร + Dependency Ratio ---
pop_df['year'] = pd.to_numeric(pop_df['year'], errors='coerce')
pop_df['Value'] = pd.to_numeric(pop_df['Value'], errors='coerce')
pop_nat = pop_df.groupby('year', as_index=False)['Value'].sum().rename(columns={'Value': 'Pop_nat'})
pop_region = pop_df.groupby(['year', 'region'], as_index=False)['Value'].sum().rename(columns={'Value': 'Pop_region'})

nat_dep = pop_nat.merge(m33_nat, on='year', how='inner')
nat_dep['Non_M33_nat'] = nat_dep['Pop_nat'] - nat_dep['M33_nat']
nat_dep['Tax_Bearer_Dependency_Ratio'] = np.where(nat_dep['M33_nat'] > 0, nat_dep['Non_M33_nat'] / nat_dep['M33_nat'], np.nan)
nat_dep = nat_dep.sort_values('year')

region_dep = pop_region.merge(m33_region, on=['year', 'region'], how='inner')
region_dep['Non_M33_region'] = region_dep['Pop_region'] - region_dep['M33_region']
region_dep['Tax_Bearer_Dependency_Ratio_region'] = np.where(region_dep['M33_region'] > 0, region_dep['Non_M33_region'] / region_dep['M33_region'], np.nan)
latest_region_year = int(region_dep['year'].max())
region_dep_latest = (
    region_dep[region_dep['year'] == latest_region_year]
    .dropna(subset=['Tax_Bearer_Dependency_Ratio_region'])
    .sort_values('Tax_Bearer_Dependency_Ratio_region', ascending=False)
    .reset_index(drop=True)
)

# --- ภาษี PIT + กำลังแรงงาน ---
if 'prov' in tax_df.columns:
    tax_df = tax_df.rename(columns={'prov': 'province'})
tax_df['amt_rcv'] = pd.to_numeric(tax_df['amt_rcv'], errors='coerce').fillna(0)
pit_df = tax_df[tax_df['tax_type'] == 1]
pit_region = pit_df.groupby(['year', 'region'], as_index=False)['amt_rcv'].sum().rename(columns={'amt_rcv': 'PIT_receipts_region'})
pit_nat = pit_df.groupby('year', as_index=False)['amt_rcv'].sum().rename(columns={'amt_rcv': 'PIT_nat'})

labor_df['year'] = pd.to_numeric(labor_df['year'], errors='coerce')
labor_df['value'] = pd.to_numeric(labor_df['value'], errors='coerce')
emp_df = labor_df[labor_df['labor_force_status'] == 'กำลังแรงงานปัจจุบัน_ผู้มีงานทำ']
emp_region = emp_df.groupby(['year', 'region'], as_index=False)['value'].sum().rename(columns={'value': 'Employed_region'})

region_fiscal = region_dep.merge(emp_region, on=['year', 'region'], how='inner').merge(pit_region, on=['year', 'region'], how='inner')
region_fiscal['Fiscal_Pressure_Index'] = np.where(
    (region_fiscal['Employed_region'] > 0) & (region_fiscal['PIT_receipts_region'] > 0) & (region_fiscal['Non_M33_region'] > 0),
    (region_fiscal['Non_M33_region'] ** 2) / (region_fiscal['Employed_region'] * region_fiscal['PIT_receipts_region']),
    np.nan
)
region_fiscal_latest_year = int(region_fiscal['year'].max())
region_fiscal_latest = (
    region_fiscal[region_fiscal['year'] == region_fiscal_latest_year]
    .dropna(subset=['Fiscal_Pressure_Index'])
    .sort_values('Fiscal_Pressure_Index', ascending=False)
    .reset_index(drop=True)
)

# --- อายุ (สำหรับข้อ 3: สังคมสูงวัย) ---
def map_age(gen):
    if gen == '0_0-15': return 'Children (0-15)'
    if gen == '11_morethanequal60': return 'Elderly (60+)'
    return 'Working Age (15-59)'
pop_df['Age_Group'] = pop_df['Generation'].apply(map_age)
pop_yearly = pop_df.groupby(['year', 'Age_Group'])['Value'].sum().unstack(fill_value=0)
for c in ['Children (0-15)', 'Working Age (15-59)', 'Elderly (60+)']:
    if c not in pop_yearly.columns: pop_yearly[c] = 0
pop_yearly['Total_pop'] = pop_yearly.sum(axis=1)
pop_yearly['Elderly_share_pct'] = np.where(pop_yearly['Total_pop'] > 0, pop_yearly['Elderly (60+)'] / pop_yearly['Total_pop'] * 100, np.nan)

nat_aging = nat_dep.merge(pit_nat, on='year', how='inner').merge(pop_yearly[['Elderly_share_pct']], left_on='year', right_index=True, how='left')
nat_aging['Tax_per_dependent'] = np.where(nat_aging['Non_M33_nat'] > 0, nat_aging['PIT_nat'] / nat_aging['Non_M33_nat'], np.nan)
valid = nat_aging.dropna(subset=['Elderly_share_pct', 'Tax_Bearer_Dependency_Ratio', 'Tax_per_dependent'])
cor_dep = valid['Elderly_share_pct'].corr(valid['Tax_Bearer_Dependency_Ratio']) if len(valid) > 1 else np.nan
cor_tax = valid['Elderly_share_pct'].corr(valid['Tax_per_dependent']) if len(valid) > 1 else np.nan
```

```python
# ข้อ 1: อัตราส่วนผู้พึ่งพิง (ตาราง + กราฟ)
first_year = int(nat_dep['year'].min())
last_year = int(nat_dep['year'].max())
dr0 = float(nat_dep.loc[nat_dep['year'] == first_year, 'Tax_Bearer_Dependency_Ratio'].iloc[0])
dr1 = float(nat_dep.loc[nat_dep['year'] == last_year, 'Tax_Bearer_Dependency_Ratio'].iloc[0])
n_years = max(1, last_year - first_year)
pct_change = (dr1 - dr0) / dr0 * 100
cagr = ((dr1 / dr0) ** (1 / n_years) - 1) * 100

display(Markdown(f"**ทั้งประเทศ** {first_year}→{last_year}: DR = {dr0:.4f} → {dr1:.4f} (เปลี่ยน {pct_change:.2f}%, CAGR ~{cagr:.3f}%/ปี)"))
cols_nat = ['year', 'Pop_nat', 'M33_nat', 'Non_M33_nat', 'Tax_Bearer_Dependency_Ratio']
display(style_pastel_table(nat_dep[cols_nat].sort_values('year').round({'Tax_Bearer_Dependency_Ratio': 4}), highlight_first=False))

top1 = region_dep_latest.iloc[0]
display(Markdown(f"**ภาคที่แรงกดดันด้านประชากรสูงสุด (ปี {latest_region_year}):** {top1['region']} (Non_M33/M33 = {top1['Tax_Bearer_Dependency_Ratio_region']:.4f})"))
cols_region = ['region', 'Tax_Bearer_Dependency_Ratio_region', 'Pop_region', 'M33_region', 'Non_M33_region']
display(style_pastel_table(region_dep_latest[cols_region].head(N_REGIONS).round(4), highlight_first=True))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
years_int = nat_dep['year'].astype(int)
dr_vals = nat_dep['Tax_Bearer_Dependency_Ratio']

# กราฟซ้าย: เส้นทั้งประเทศ
ax1.plot(years_int, dr_vals, marker='o', markersize=8, color='#4c72b0', linewidth=2, label='Dependency Ratio')
ax1.set_xticks(years_int)
ax1.set_xticklabels([str(int(v)) for v in years_int], rotation=0)
ax1.set_xlabel('ปี', fontsize=CHART['label_fontsize'], fontweight='normal')
ax1.set_ylabel('Non_M33 / M33', fontsize=CHART['label_fontsize'], fontweight='normal')
ax1.set_title('Dependency Ratio (Non_M33/M33) ทั้งประเทศ', fontsize=CHART['title_fontsize'], pad=CHART['title_pad'], fontweight='bold')
ax1.set_ylim(0, dr_vals.max() * 1.2)
ax1.grid(True, axis='y', linestyle=CHART['grid_ls'], alpha=CHART['grid_alpha'])
for x, y in zip(years_int, dr_vals):
    ax1.annotate(f'{y:.2f}', (x, y), textcoords='offset points', xytext=(0, 10), ha='center', fontweight='bold', color='#4c72b0', fontsize=11)
ax1.legend(loc='upper right', frameon=True)

# กราฟขวา: แท่งรายภาค + เส้น + เลขกำกับ
plot_r = region_dep_latest.head(N_REGIONS).sort_values('Tax_Bearer_Dependency_Ratio_region', ascending=True)
regions_r = plot_r['region'].astype(str)
vals_r = plot_r['Tax_Bearer_Dependency_Ratio_region']

ax2.barh(regions_r, vals_r, color='#4c72b0', edgecolor='white', height=0.7, label='Dependency Ratio (Bar)')
ax2.plot(vals_r, regions_r, color='#dd8452', marker='o', markersize=7, linewidth=2, label='Dependency Ratio (Line)')
for y, x in zip(regions_r, vals_r):
    ax2.annotate(f'{x:.2f}', (x, y), textcoords='offset points', xytext=(8, 0), va='center', fontweight='normal', color='#4c72b0', fontsize=10)

ax2.set_xlabel('Non_M33_region / M33_region', fontsize=CHART['label_fontsize'], fontweight='normal')
ax2.set_ylabel('Fiscal Pressure Index', fontsize=CHART['label_fontsize'], fontweight='normal')
ax2.set_title(f'Dependency Ratio รายภาค (ปี {latest_region_year})', fontsize=CHART['title_fontsize'], pad=CHART['title_pad'], fontweight='bold')
ax2.grid(True, axis='x', linestyle=CHART['grid_ls'], alpha=CHART['grid_alpha'])
ax2.legend(loc='lower right', frameon=True)
plt.tight_layout()
#plt.savefig('.png')
plt.show()
```

## 2. แรงกดดันทางการคลัง: ภาคไหนเสี่ยงที่สุด?

เมื่อคิดรวม **ผู้พึ่งพิง + กำลังแรงงาน + รายได้ภาษี (PIT)**  
**Fiscal Pressure Index = Non_M33² ÷ (Employed × PIT)** — ยิ่งสูง = แรงกดดันสูง (ภาคนั้นมีผู้พึ่งพิงมาก แต่ฐานแรงงาน/ภาษีรองรับจำกัด)


```python
# Insight-driven scatter: เชื่อม Dependency Ratio กับ Fiscal Pressure Index (รายภูมิภาค)

# 1) รวมตัวชี้วัด “รายภูมิภาค” จากข้อมูลที่มีอยู่แล้ว (ไม่สร้าง DataFrame ใหม่จากค่าคงที่)
df_pair = region_dep_latest[['region', 'Tax_Bearer_Dependency_Ratio_region']].merge(
    region_fiscal_latest[['region', 'Fiscal_Pressure_Index']],
    on='region',
    how='inner'
)

# 2) สเกลแกน Y เพื่อให้อ่านง่าย (เพราะ Fiscal Pressure Index มีค่าน้อยมาก)
df_pair['Fiscal_Pressure_x1e6'] = df_pair['Fiscal_Pressure_Index'] * 1e6

x = df_pair['Tax_Bearer_Dependency_Ratio_region']
y = df_pair['Fiscal_Pressure_x1e6']

# 3) หา trend line แบบ linear (y = slope*x + intercept) และ correlation
x_mean = x.mean()
y_mean = y.mean()

slope = ((x - x_mean) * (y - y_mean)).sum() / ((x - x_mean) ** 2).sum()
intercept = y_mean - slope * x_mean

corr = x.corr(y)

# จุดสำหรับเส้นแนวโน้ม (ทำแบบ list เพื่อความเข้ากันกับสไตล์ notebook)
x_min, x_max = float(x.min()), float(x.max())
x_line = [x_min + (x_max - x_min) * i / 100 for i in range(101)]
y_line = [slope * xx + intercept for xx in x_line]

# 4) วาดกราฟ scatter (แกน X=Dependency Ratio, แกน Y=Fiscal Pressure)
fig, ax = plt.subplots(figsize=(11, 6))

default_color = CHART.get('line_blue', '#4c72b0')
central_color = '#d62728'  # outlier
bkk_color = '#2ca02c'       # low-low zone

for _, r in df_pair.iterrows():
    region = str(r['region'])
    xi = float(r['Tax_Bearer_Dependency_Ratio_region'])
    yi = float(r['Fiscal_Pressure_x1e6'])

    if region == 'ภาคกลาง':
        color = central_color
        size = 140
    elif region == 'กรุงเทพมหานครและปริมณฑล' or 'กรุงเทพ' in region:
        color = bkk_color
        size = 120
    else:
        color = default_color
        size = 90

    ax.scatter(
        xi,
        yi,
        s=size,
        color=color,
        edgecolor='white',
        linewidth=1.0,
        alpha=0.95
    )

    # ใส่ชื่อภูมิภาคกำกับทุกจุด
    ax.annotate(
        region,
        (xi, yi),
        textcoords='offset points',
        xytext=(7, 6),
        ha='left',
        fontsize=10
    )

# trend line
ax.plot(
    x_line,
    y_line,
    linestyle='--',
    linewidth=2,
    color='gray',
    label=f'Trend line (corr={corr:.3f})'
)

# 5) Annotation สั้นๆ เพื่อเล่า insight
central_row = df_pair[df_pair['region'] == 'ภาคกลาง'].iloc[0]
bkk_row = df_pair[df_pair['region'].astype(str).str.contains('กรุงเทพ', na=False)].iloc[0]

ax.annotate(
    'Outlier: ภาคกลาง\nDR สูง & Fiscal สูง',
    (float(central_row['Tax_Bearer_Dependency_Ratio_region']), float(central_row['Fiscal_Pressure_x1e6'])),
    textcoords='offset points',
    xytext=(-45, 35),
    fontsize=11,
    fontweight='bold',
    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.9, edgecolor=central_color)
)

ax.annotate(
    'โซนต่ำทั้งคู่: กรุงเทพฯ',
    (float(bkk_row['Tax_Bearer_Dependency_Ratio_region']), float(bkk_row['Fiscal_Pressure_x1e6'])),
    textcoords='offset points',
    xytext=(10, -25),
    fontsize=11,
    fontweight='bold',
    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.9, edgecolor=bkk_color)
)

# 6) จัดแต่งกราฟ
ax.set_title(
    'Dependency Ratio vs Fiscal Pressure Index (รายภูมิภาค, ปีล่าสุด)',
    fontsize=CHART.get('title_fontsize', 14),
    pad=CHART.get('title_pad', 15),
    fontweight='bold'
)
ax.set_xlabel('Dependency Ratio (Non_M33 / M33)', fontsize=CHART.get('label_fontsize', 12), fontweight='normal')
ax.set_ylabel('Fiscal Pressure Index (x 1e6)', fontsize=CHART.get('label_fontsize', 12), fontweight='normal')

ax.grid(True, linestyle=CHART.get('grid_ls', '--'), alpha=CHART.get('grid_alpha', 0.35))
ax.legend(loc='upper left', frameon=True)

plt.tight_layout()
plt.show()

print('สรุปสั้นๆ: กราฟนี้แสดงว่าเมื่อ Dependency Ratio สูงขึ้น มักสัมพันธ์กับ Fiscal Pressure Index ที่สูงขึ้น โดยมี outlier เด่นคือภาคกลาง และโซนต่ำเด่นคือกรุงเทพฯ')
```
